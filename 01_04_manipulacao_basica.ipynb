## 1. Manipulação de Pixels

````python
# Lê imagem como escala de cinza
mnist_3_img = cv2.imread('imagens/MNIST_3_18x18.png', cv2.IMREAD_GRAYSCALE)

# Apresenta imagem
plt.figure(figsize=(4, 4))
plt.imshow(mnist_3_img, cmap='gray');
````

### 1.1. Acessando Pixels Individualmente

Como vimos anteriormente, os valores de intensidade dos pixels pode ser feito através do acesso ao array da imagem. Lembrando que a notação é `matriz[linha, coluna]` e que o índice começa em `0`.

````python
print(mnist_3_img[3, 10])
print(mnist_3_img[12, 2])
print(mnist_3_img[10, 12])
````

### 1.2. Modificando Pixels

Podemos modificar de diversas formas, como mostrado abaixo.

````python
# Criando uma cópia
mnist_3_img_copy = mnist_3_img.copy()

# Modificando pixels individualmente
mnist_3_img_copy[2, 2] = 100
mnist_3_img_copy[2, 3] = 125
mnist_3_img_copy[3, 2] = 150
mnist_3_img_copy[3, 3] = 175

# Modificando pixels através de slicings nos arrays NumPy
mnist_3_img_copy[0:18, 0] = 150
mnist_3_img_copy[0:18, 17] = 150
mnist_3_img_copy[0, 0:18] = 150
mnist_3_img_copy[17, 0:18] = 150

# Apresentando as intensidades
print(mnist_3_img_copy)

# Apresentando a imagem
plt.figure(figsize = (4, 4))
plt.imshow(mnist_3_img_copy, cmap='gray');
````

## 2. Recortando Imagens

Recortar uma imagem não requer nenhuma função especial do OpenCV. Pode ser feito selecionando a região específica da imagem.

````python
tucanos_img = cv2.imread('imagens/tucanos.jpg', cv2.IMREAD_COLOR)

plt.figure(figsize = (8, 8))
plt.imshow(tucanos_img[:, :, ::-1]);
````

### Recortando região da imagem

````python
cropped_img = tucanos_img[1450:-1, 330:1000]
plt.imshow(cropped_img[:, :, ::-1]);
````

## 3. Redimensionando Imagens

A função **`resize()`** redimensiona a imagem `src`, reduzindo ou ampliando suas dimensões para o tamanho especificado. O tamanho e o tipo da imagem de saída são determinados com base em `src`, `dsize`, `fx` e `fy`.

---

### <font color="green">Sintaxe da Função</font>
```python
dst = cv2.resize(src, dsize[, dst[, fx[, fy[, interpolation]]]])
```

`dst`: imagem de saída. Ela terá o tamanho definido por `dsize` (quando este for diferente de zero) ou o tamanho calculado a partir de `src.size()`, `fx` e `fy`. O tipo de dados de `dst` será o mesmo de `src`.

A função possui **2 argumentos obrigatórios**:

1. `src`: imagem de entrada.
2. `dsize`: tamanho da imagem de saída.

Os argumentos opcionais mais utilizados incluem:

1. `fx`: fator de escala no eixo horizontal (largura).
2. `fy`: fator de escala no eixo vertical (altura).

A imagem de saída terá o tamanho especificado por `dsize` (quando este for diferente de zero) ou um tamanho calculado a partir do tamanho original da imagem (`src.size()`) e dos fatores de escala `fx` e `fy`. O tipo de dados da imagem de saída (`dst`) permanece o mesmo da imagem de entrada (`src`).

Se os parâmetros `fx` e `fy` forem fornecidos, o argumento `dsize` deve ser definido como `None`, já que o novo tamanho será calculado automaticamente a partir dos fatores de escala informados.


### 3.1. Método 1: Especificando Fator de Scala usando fx e fy

````python
# Redimensiona imagem na escala de 2
resized_cropped_img_2x = cv2.resize(cropped_img, None, fx=2, fy=2)

# Apresenta imagem
plt.imshow(resized_cropped_img_2x[:, :, ::-1]);
````

### 3.2. Método 2: Especificando Tamanho Exato da Imagem

````python
# Configura tamanho desejado
desired_width  = 400
desired_height = 200
dim = (desired_width, desired_height)

# Redimensiona e apresenta
resized_cropped_img = cv2.resize(cropped_img, dsize=dim, interpolation=cv2.INTER_AREA)
plt.figure(figsize=(6, 6))
plt.imshow(resized_cropped_img[:, :, ::-1]);
````

### 3.3. Método 3: Redimensionando Mantendo Proporção

````python
# Usando dsize (método 2)
desired_width = 200
aspect_ratio = cropped_img.shape[0] / cropped_img.shape[1]
desired_height = int(desired_width * aspect_ratio)
dim = (desired_width, desired_height)

# Redimensionando e apresentando
resized_cropped_img = cv2.resize(cropped_img, dsize=dim, interpolation=cv2.INTER_AREA)
plt.figure(figsize=(6, 6))
plt.imshow(resized_cropped_img[:, :, ::-1]);
````

## 4. Espelhando Imagens

A função **`flip()`** realiza o espelhamento (*flip*) de um array ou imagem de três maneiras diferentes (os índices de linhas e colunas são baseados em zero).

### <font color="green">Sintaxe da Função</font>
```python
dst = cv2.flip(src, flipCode[, dst])
```

`dst`: array (ou imagem) de saída com o mesmo tamanho e o mesmo tipo de dados de `src`.

A função possui **2 argumentos obrigatórios**:

1. `src`: imagem de entrada.
2. `flipCode`: parâmetro que define a forma de espelhamento da imagem:
   * **0**: espelhamento em torno do eixo **x** (inverte a imagem verticalmente, de cima para baixo).
   * **Valor positivo** (por exemplo, **1**): espelhamento em torno do eixo **y** (inverte a imagem horizontalmente, da esquerda para a direita).
   * **Valor negativo** (por exemplo, **-1**): espelhamento em torno dos dois eixos, realizando simultaneamente a inversão horizontal e vertical.

Em termos práticos:

* `flipCode = 0` → *flip vertical* (topo ↔ base).
* `flipCode = 1` → *flip horizontal* (esquerda ↔ direita).
* `flipCode = -1` → *flip horizontal e vertical* ao mesmo tempo.

Essa função é amplamente utilizada em processamento de imagens e visão computacional para realizar **aumento de dados (*data augmentation*)**, correção de orientação e geração de diferentes perspectivas de uma mesma imagem.


````python
# Espelha (flip) dos 3 jeitos
tucanos_img_flipped_horz = cv2.flip(cropped_img, 1)
tucanos_img_flipped_vert = cv2.flip(cropped_img, 0)
tucanos_img_flipped_both = cv2.flip(cropped_img, -1)

# Apresenta imagens
plt.figure(figsize = [12, 5])
plt.subplot(141); plt.imshow(tucanos_img_flipped_horz[:, :, ::-1]) 
plt.title('Horizontal')
plt.subplot(142); plt.imshow(tucanos_img_flipped_vert[:, :, ::-1])
plt.title('Vertical')
plt.subplot(143); plt.imshow(tucanos_img_flipped_both[:, :, ::-1]) 
plt.title('Ambos')
plt.subplot(144); plt.imshow(cropped_img[:, :, ::-1])            
plt.title('Original');
````
