## 2. Lendo Imagens com OpenCV

Agora vamos usar a função `imread()` do OpenCV para ler as imagens e armazená-las em arrays 2D NumPy.

---

Ler imagens no OpenCV é bem simples. Vamos usar a função `imread()` que nos permite ler diferentes tipos de arquivos de imagem (JPG, PNG, etc). Você pode lê-las como imagens em escala de cinza ou coloridas. Você também pode ler imagens que contenham o canal alpha.

### <font color="green">Sintaxe da Função</font>
``` python
retval = cv2.imread(filename[, flags])
```

`retval`: É a imagem caso tenha carregado com sucesso. Do contrário, será `None`. Isso pode ocorrer caso o nome do arquivo esteja errado ou corrompido.

Essa função possui apenas **1 argumento requerido** e **1 flag opcional**:

1. `filename`: Pode ser tanto um caminho absoluto ou relativo. É um argumento obrigatório.
2. `flags`: Flags (opções) usadas para ler uma imagem em um formato específico (por exemplo, em escala de cinza, colorido, com canal alpha). Esse é um argumento opcional e seu valor padrão é `cv2.IMREAD_COLOR` ou `1` que carrega a imagem como uma imagem colorida.

Antes de apresentarmos alguns exemplos, vamos olhar alguns `flags` disponíveis.

1. **`cv2.IMREAD_GRASCALE`** ou **`0`**: Carrega a imagem como escala de cinza.
2. **`cv2.IMREAD_COLOR`** ou **`1`**: Carrega como uma imagem colorida. Qualque transparência será ignorada. Esse é o flag padrão.
3. **`cv2.IMREAD_UNCHANGED`** ou **`-1`**: Carrega a imagem com seus canais originais, que pode incluir o canal alpha.

### 2.1. Lendo Imagens e Apresentando os Dados de Pixel

Vamos ler a imagem 18x18 em escala de cinza usando `imread()` e apresentando as intensidades dos pixels. Os dados são armazenados como inteiros 8-bit sem sinal (`uint8`), então os valores de intensidade vão de 0 a 255, onde 0 representa a cor preta pura e 255 o branco puro.

````python
filename = 'imagens/MNIST_3_18x18.png'
# Lendo a imagem como escala de cinza
pb_img = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)

# Apresenta os valores dos pixels como um array 2D NumPy
# Cada valor de pixel está no range de 0 a 255
print(pb_img)
````

### 2.2. Obtendo Atributos da Imagem

````python
# Imprime o tamanho da imagem.
print('O tamanho da imagem é', pb_img.shape)

# Imprime o tipo de dado da imagem.
print('O tipo de dado da imagem é', pb_img.dtype)
````

## 3. Coordenadas da Imagem

O sistema de coordenada usado para representar os dados da imagem é um sistema cartesiano onde a origem está no canto superior esquerdo da imagem. A largura é medido ao longo do eixo-x e a altura ao longo do eixo-y. Quando o dado está armazenado em uma matriz 2D, o primeiro índice representa as linhas (eixo-y) e o segundo índice representa as colunas (eixo-x), ou seja `[y, x]`. Esse sistema de coordenada mantém a convenção onde o eixo-x representa a direção horizontal e o eixo-y a direção vertical.

Muita atenção, pois algumas funções do OpenCV adotam a notação padrão da matemática `(x, y)`.

### <font color="red">Atenção</font>

* Acessando os dados de um pixel em um array NumPy: `[y, x]`
* Algumas funções do OpenCV: `(x, y)`
* Índices em Python começam do `0` e não do `1`.

````python
print(pb_img)
````

````python
# Acessando a pontinha inferior do 3
pb_img[12, 2]
````

````python
# Acessando imagem com dimensões diferentes
filename = 'imagens/lagoa-da-pampulha.webp'
lagoa_img = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)

# Acessando atributos
print("Dimensões:", lagoa_img.shape)
print("Altura:", lagoa_img.shape[0])
print("Largura:", lagoa_img.shape[1])
````

## 4. Apresentando os Dados da Imagem como uma "Imagem" usando Matplotlib

Uma vez que os dados da imagem estão armazenados em um array 2D NumPy, podemos apresentá-los usando a função `imshow()` do matplotlib. Existem alguns pontos de observação.

1. A representação de cor é inesperada
2. A imagem apresentada é uma representação matemática. A imagem é apenas 18x18 pixels como indicado nos eixos, mas o tamanho físico no browser é muito maior. Isso pode ser conveniente, pois podemos mostrar imagens de diversos tamanhos em uma escala de visão adequada. O tamanho da plotagem é configurável usando `plt.figure(figsize=[width, height])`

````python
# Apresentando a imagem
plt.imshow(pb_img);
````

### O que aconteceu com as cores?
Mesmo que a imagem tenha sido lida como uma imagem em escala de cinza, não será necessariamente apresentada como escala de cinza usando `imshow()`. Matplotlib usar diferentes mapeamentos de cor para interpretar intensidades dos pixels. Para renderizar adequadamente essa imagem como escala de cinza, nós precisamos especificar o mapa de cores. Como mostrado abaixo, podemos especificar o mapa de cores para escala de cinza passando um segundo argumento (opcional).

````python
# Ajustando o mapa de cores para escala de cinza
plt.imshow(pb_img, cmap='gray');
````

## 5. Usando Outra Imagem em Escala de Cinza

Agora vamos carregar uma nova imagem em escala de cinza, mas sem forçar carregar com `cv2.IMREAD_GRAYSCALE`. Se não carregar dessa forma, também não precisamos configurar o argumento `cmap` da função `imshow()` do matplotlib.

````python
# Acessando imagem, mas não forçando carregar em escala de cinza
filename = 'imagens/lagoa-da-pampulha.webp'
lagoa_img = cv2.imread(filename)

# Configurando tamanho da figura
plt.figure(figsize=[10, 10])
# Apresentação sem configurar mapa de cores, pois carregamos a imagem com todos os canais
plt.imshow(lagoa_img)
# Dando um título para a figura
plt.title('Lagoa da Pampulha', fontsize=16);
````

## 6. Salvando Imagens

É bem simples de salvar imagens usando OpenCV usando a função `imwrite()`. Essa função salva a imagem no arquivo especificado. O formato da imagem é escolhido baseado na extensão do arquivo.

### <font color="green">Sintaxe da Função</font>
``` python
cv2.imwrite(filename, img[, params])
```

A função possui **2 argumentos requeridos**:

1. `filename`: Pode ser o caminho absoluto ou relativo.
2. `img`: Imagem para ser salva

### Lendo e apresentando a imagem

````python
lagoa_img = cv2.imread('imagens/lagoa-da-pampulha.webp')
plt.figure(figsize=[10, 10])
plt.imshow(lagoa_img)
plt.title('lagoa-da-pampulha.webp', fontsize=16);
````

### Salvando a imagem como PNG

````python
cv2.imwrite('output/lagoa-da-pampulha.png', lagoa_img)
Image('output/lagoa-da-pampulha.png', width=300)
````

### Salvando a imagem como JPEG alterando qualidade

Existem diversos parâmetros que podem ser configurados ao salvar uma imagem. No exemplo abaixo, configuraremos a qualidade do arquivo JPEG para outro que quisermos (padrão é 95).

````python
# Salvando arquivo como JPEG de baixa qualidade
cv2.imwrite('output/lagoa-da-pampulha.jpg', lagoa_img, (cv2.IMWRITE_JPEG_QUALITY, 10))
Image('output/lagoa-da-pampulha.jpg', width=300)
````
