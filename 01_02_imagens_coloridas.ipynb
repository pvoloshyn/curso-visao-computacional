## 1. Lendo e Apresentando Imagens Coloridas

Até agora, nós só lidamos com arquivos em escala de cinza. Vamos agora explorar fazer o mesmo com imagens coloridas.

### 1.1. Exemplo: Imagem Colorida formato JPEG

Vamos começar com uma imagem colorida simples em formato JPEG.

````python
# Será essa a imagem que apresentaremos
Image(filename='imagens/puc-minas.jpg')
````

### Lendo imagem JPEG e apresentando suas dimensões

Imagens coloridas são tipicamente representadas usando 3 canais de cor separados. Os canais de cor usados para representar dependem do seu espaço de cor. O espaço de cor mais comum é o RGB, que contém os canais R (vermelho), G (verde) e B (azul).

````python
# Lendo a imagem
filename = 'imagens/puc-minas.jpg'
puc_img = cv2.imread(filename, cv2.IMREAD_COLOR)

# Apresentando suas dimensões
print("Dimensões da imagem:", puc_img.shape)
````

### Apresentando a imagem

````python
plt.figure(figsize=(10, 10))
plt.imshow(puc_img);
````

### O que aconteceu com as cores?

As cores apresentadas diferem da imagem original. Isso ocorre, pois a biblioteca matplotlib espera que a imagem esteja no formato RGB, porém o OpenCV armazena as imagens no formato BGR. Diante disso, para corrigir, precisamos inverter a ordem dos canais.

### Usando cvtColor() para trocar de espaço de cor

Felizmente, existe uma função do OpenCV para converter um espaço de cor para outro. A função é a `cvtColor()`.

---
### <font color="green">Sintaxe da Função</font>
``` python
dst = cv2.cvtColor(src, code)
```

`dst`: É a imagem de saída do mesmo tamanho e profundidade de `src`

Essa função possui **2 argumentos requeridos**:

1. `src`: Imagem de entrada.
2. `code`: Código de conversão de espaços de cor


### Convertendo de BGR para RGB

````python
# Converte espaço de cor de BGR para RGB
puc_img = cv2.cvtColor(puc_img, cv2.COLOR_BGR2RGB)

# Apresenta imagem
plt.figure(figsize=(10, 10))
plt.imshow(puc_img);
````

### 1.2. Exemplo: Imagem Colorida formato PNG

Agora vamos usar uma imagem colorida e com transparência em formato PNG.

````python
# Será essa a imagem que apresentaremos
Image(filename='imagens/puc-minas-logo.png', width=300)
````

### Lendo e apresentando a imagem PNG

Vamos usar a mesma abordagem que usamos para imagens JPG para ver como irá se comportar.

````python
# Lendo a imagem
filename = 'imagens/puc-minas-logo.png'
logo_img = cv2.imread(filename, cv2.IMREAD_COLOR)

# Apresentando suas dimensões
print("Dimensões da imagem:", logo_img.shape)

# Converte espaço de cor de BGR para RGB
logo_img = cv2.cvtColor(logo_img, cv2.COLOR_BGR2RGB)

# Apresenta imagem
plt.figure(figsize=(10, 10))
plt.imshow(logo_img);
````

### O que aconteceu?

Duas coisas ocorreram aqui:

1. Usamos a flag `cv2.IMREAD_COLOR` que descarta o canal de transparência
2. Trabalhamos apenas no espaço de cor RGB. Precisamos lidar com o RGBA.

### Lendo e apresentando PNG de forma correta

Vamos usar a flag `cv2.IMREAD_UNCHANGED` para não descartar nenhum canal e usar o código `cv2.COLOR_BGRA2RGBA` para considerar o canal de transparência na conversão de espaços de cor.

````python
# Lendo a imagem, não descartando nada
filename = 'imagens/puc-minas-logo.png'
logo_img = cv2.imread(filename, cv2.IMREAD_UNCHANGED)

# Apresentando suas dimensões
print("Dimensões da imagem:", logo_img.shape)

# Converte espaço de cor de BGRA para RGBA
logo_img = cv2.cvtColor(logo_img, cv2.COLOR_BGRA2RGBA)

# Apresenta imagem
plt.figure(figsize=(10, 10))
plt.imshow(logo_img);
````

## 3. Separando e Mesclando Canais de Cor

Vamos explorar agora como separar e mesclar canais de cor usando as funções `split()` e `merge()` do OpenCV.

---

`split()` Divide um array multi-canal em diversos arrays de único-canal.

`merge()` Mescla diversos arrays para compor um único array multi-canal. Todas as matrizes de entrada precisam ter o mesmo tamanho.

````python
# Lendo a imagem
filename = 'imagens/puc-minas.jpg'
bgr_img = cv2.imread(filename, cv2.IMREAD_COLOR)

# Separando os canais de cor
b, g, r = cv2.split(bgr_img)

# Apresenta os canais de cor separadamente
plt.figure(figsize = [20, 10])
plt.subplot(141); plt.imshow(r, cmap='gray'); plt.title('R')
plt.subplot(142); plt.imshow(g, cmap='gray'); plt.title('G')
plt.subplot(143); plt.imshow(b, cmap='gray'); plt.title('B')

# Combina todos os canais de cor
merged_img = cv2.merge((r, g, b))

# Apresenta imagem mesclada
plt.subplot(144)
plt.imshow(merged_img)
plt.title('Saída Mesclada');
````

## 4. Convertendo para Diferentes Espaços de Cor

### 4.1. Conceitos sobre Espaços de Cor

Em termos simples, um **espaço de cores** é uma organização específica das cores que normalmente representa o conjunto de todas as cores que os seres humanos conseguem perceber. Um **modelo de cores** é uma construção matemática que define como especificar cores dentro desse espaço utilizando uma combinação única de números (geralmente três ou quatro valores que representam as contribuições relativas dos componentes de cor). Um modelo de cores pode ser entendido como uma forma matemática de navegar por um espaço de cores. No entanto, é muito comum que o termo **“espaço de cores”** seja usado para se referir tanto ao modelo de cores quanto ao mapeamento específico desse modelo para um espaço de cores absoluto.

Como introdução aos espaços de cores, consideraremos dois modelos amplamente utilizados: o espaço de cores **RGB** (*Red, Green, Blue* — Vermelho, Verde e Azul) e o espaço de cores **HSV** (*Hue, Saturation, Value* — Matiz, Saturação e Valor). Ambos utilizam um sistema de coordenadas tridimensional para especificar os componentes de cor que formam uma combinação única de valores e, consequentemente, uma cor única. Esses componentes também são chamados de **canais de cor**.

Como as imagens coloridas geralmente são representadas por três canais de cor utilizando inteiros sem sinal de 8 bits para cada canal, os componentes individuais podem assumir valores no intervalo de **0 a 255**. Dessa forma, é possível representar **16,77 milhões de cores únicas** em qualquer um desses espaços de cores (**256 × 256 × 256**).

Nos exemplos a seguir, trabalharemos com imagens coloridas nos espaços de cores **RGB** e **HSV**.

### 4.2 Mudando para o Espaço de Cor HSV

Nós também podemos usar a função `cvtColor()` para converter de um espaço de cor para outro. No exemplo abaixo, vamos converter a imagem para o espaço de cor HSV e dividir os canais.

````python
# Convertendo de BGR para HSV
hsv_img = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2HSV)

# Separando os canais de cor
h, s, v = cv2.split(hsv_img)

# Apresenta os canais de cor separadamente
plt.figure(figsize = [20, 10])
plt.subplot(141); plt.imshow(h, cmap='gray'); plt.title('H')
plt.subplot(142); plt.imshow(s, cmap='gray'); plt.title('S')
plt.subplot(143); plt.imshow(v, cmap='gray'); plt.title('V')

# Apresentando a imagem original. Note que invertemos os canais de cor de outra forma
plt.subplot(144); plt.imshow(bgr_img[:, :, ::-1]); plt.title('Original');
````

### 4.3. Modificando Canais de Cor

````python
h_novo = h + 30
hsv_img_merged = cv2.merge((h_novo, s, v))
rgb_img_merged = cv2.cvtColor(hsv_img_merged, cv2.COLOR_HSV2RGB)

# Apresenta os canais de cor
plt.figure(figsize = [20, 10])
plt.subplot(141); plt.imshow(h_novo, cmap='gray'); plt.title('H')
plt.subplot(142); plt.imshow(s, cmap='gray'); plt.title('S')
plt.subplot(143); plt.imshow(v, cmap='gray'); plt.title('V')

# Apresentando a imagem original. Note que invertemos os canais de cor de outra forma
plt.subplot(144); plt.imshow(rgb_img_merged); plt.title('Modificado');
````

## 5. Salvando Imagens Coloridas

Já vimos como salvar imagens, mas em escala de cinza. A função e parâmetros não muda, porém, a mesma atenção que temos com os canais de cor RGB e BGR deve se manter.

````python
# Lendo no formato BGR
bgr_img = cv2.imread('imagens/puc-minas.jpg', cv2.IMREAD_COLOR)
# Convertendo para RGB
rgb_img = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(5, 5))
plt.imshow(rgb_img);
````

````python
# Salvando como RGB
cv2.imwrite('output/rgb.jpg', rgb_img)
Image('output/rgb.jpg', width=300)
````

````python
# Salvando como BGR
cv2.imwrite('output/bgr.jpg', bgr_img)
Image('output/bgr.jpg', width=300)
````
